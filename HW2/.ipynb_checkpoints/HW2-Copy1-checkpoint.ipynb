```python
domain=[]
best_profit=0
best_comb=np.zeros(n)
#
for i in range(2**n):
    # '0b'
    binary_string = bin(i)[2:]
    comb = [int(bit) for bit in binary_string.zfill(n)]
    if np.dot(comb,weight)<= capacity:
        domain.append(comb)
        if best_profit<=np.dot(comb,profit):
            best_profit = np.dot(comb,profit)
            best_comb = comb

print(best_comb,np.dot(best_comb,weight),best_profit)
```

```python
def neighbour(a,dis):#adis
    neigh=[]
    for i in range(1,dis+1):#1,dis
        for j in list(combinations(range(len(a)),i)):#
            acopy=a.copy()
            for k in j:#
                acopy[k]=(acopy[k]+1)%2
            if np.dot(acopy,weight)<capacity: #capacity
                neigh.append(acopy)
    return neigh
```
